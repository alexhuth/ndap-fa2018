```python
%config InlineBackend.figure_format = 'retina'
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
ls # lists the stuff in this directory
```

```python
data = np.load('test_data.npz')
data.files
```

```python
perf1 = data['perf1']
perf2 = data['perf2']
print(perf1.shape)
print(perf2.shape)
```

```python
print(perf1.max(), perf1.min())
print(perf2.max(), perf2.min())
```

```python
# plotting each separately
# but these datasets aren't ordered, so this is silly!
plt.plot(perf1)
plt.plot(perf2)
```

```python
# plotting a comparison
# connecting the points with lines doesn't make any sense
# and it looks like hot garbage
plt.plot(perf1, perf2)
```

```python
# plotting points is better!
# but notice the x- and y-axes. they are stretched weirdly!
plt.plot(perf1, perf2, '.')
```

```python
# first let's make the plot square instead of rectangular
# but the scaling on each axis is still different
plt.figure(figsize=(6,6)) # create a figure that is 6" by 6" (square)
plt.plot(perf1, perf2, '.')
```

```python
# here we even out the axes so that they are scaled the same
# this tells a pretty different story!
plt.figure(figsize=(6,6))
plt.plot(perf1, perf2, '.')
plt.xlim(-0.06, 0.35)
plt.ylim(-0.06, 0.35)
```

```python
# we can add a diagonal (x=y) line so that it's easy to see which points
# have perf1 > perf2 and vice versa
plt.figure(figsize=(6,6))
plt.plot(perf1, perf2, '.')
plt.plot([-0.06, 0.35], [-0.06, 0.35], '-')
plt.xlim(-0.06, 0.35)
plt.ylim(-0.06, 0.35)
```

```python
# finally we add some labels to the x- and y-axes so that the plot is easy to read
# this looks like a real plot! this is the good stuff
plt.figure(figsize=(6,6))
plt.plot(perf1, perf2, '.')
plt.plot([-0.06, 0.35], [-0.06, 0.35], '-')
plt.xlim(-0.06, 0.35)
plt.ylim(-0.06, 0.35)
plt.xlabel('Model performance 1')
plt.ylabel('Model performance 2')
```
